```python
# 1.Yearly Breakdown of Stops and Arrests by Country (Using Subquery and Window Functions)

query = """
SELECT 
    country_name,
    stop_year,
    total_stops,
    total_arrests,
    SUM(total_arrests) OVER (PARTITION BY country_name ORDER BY stop_year) AS cumulative_arrests
FROM (
    SELECT 
        country_name,
        YEAR(stop_date) AS stop_year,
        COUNT(*) AS total_stops,
        SUM(CASE WHEN is_arrested = 1 THEN 1 ELSE 0 END) AS total_arrests
    FROM traffic_stops
    GROUP BY country_name, YEAR(stop_date)
) AS yearly_data
ORDER BY country_name, stop_year;
"""

# Execute and fetch
mycursor.execute(query)
results = mycursor.fetchall()

# Setup PrettyTable
table = PrettyTable()
table.field_names = ["Country", "Year", "Total Stops", "Total Arrests", "Cumulative Arrests"]

for row in results:
    table.add_row(row)

# Print the formatted table
print(table)
```

```python
# 2.Driver Violation Trends Based on Age and Race (Join with Subquery)

query = """
SELECT 
    CASE 
        WHEN driver_age BETWEEN 18 AND 25 THEN '18-25'
        WHEN driver_age BETWEEN 26 AND 35 THEN '26-35'
        WHEN driver_age BETWEEN 36 AND 45 THEN '36-45'
        WHEN driver_age BETWEEN 46 AND 60 THEN '46-60'
        WHEN driver_age > 60 THEN '60+'
        ELSE 'Unknown'
    END AS age_group,
    driver_race,
    violation,
    COUNT(*) AS violation_count
FROM traffic_stops
WHERE driver_age IS NOT NULL
GROUP BY age_group, driver_race, violation
ORDER BY age_group, driver_race, violation_count DESC;
"""

# Execute and fetch
mycursor.execute(query)
results = mycursor.fetchall()

# Display with PrettyTable
table = PrettyTable()
table.field_names = ["Age Group", "Driver Race", "Violation", "Violation Count"]

for row in results:
    table.add_row(row)

print(table)
```

```python
# 3.Time Period Analysis of Stops (Joining with Date Functions)

query = """
SELECT 
    t.country_name,
    CASE 
        WHEN HOUR(t.stop_time) BETWEEN 5 AND 11 THEN 'Morning'
        WHEN HOUR(t.stop_time) BETWEEN 12 AND 16 THEN 'Afternoon'
        WHEN HOUR(t.stop_time) BETWEEN 17 AND 20 THEN 'Evening'
        WHEN HOUR(t.stop_time) >= 21 OR HOUR(t.stop_time) <= 4 THEN 'Night'
        ELSE 'Unknown'
    END AS time_period,
    COUNT(*) AS stop_count
FROM traffic_stops t
WHERE t.stop_time IS NOT NULL
GROUP BY t.country_name, time_period
ORDER BY t.country_name, time_period;
"""

# Execute and fetch results
mycursor.execute(query)
results = mycursor.fetchall()

# Display results using PrettyTable
table = PrettyTable()
table.field_names = ["Country", "Time Period", "Stop Count"]

for row in results:
    table.add_row(row)

print(table)
```

```python
# 4.Correlation Between Age, Violation, and Stop Duration (Subquery)

query = """
SELECT 
    age_group,
    violation,
    stop_duration,
    COUNT(*) AS count
FROM (
    SELECT 
        CASE 
            WHEN driver_age BETWEEN 18 AND 25 THEN '18-25'
            WHEN driver_age BETWEEN 26 AND 35 THEN '26-35'
            WHEN driver_age BETWEEN 36 AND 45 THEN '36-45'
            WHEN driver_age BETWEEN 46 AND 60 THEN '46-60'
            WHEN driver_age > 60 THEN '60+'
            ELSE 'Unknown'
        END AS age_group,
        violation,
        stop_duration
    FROM traffic_stops
    WHERE driver_age IS NOT NULL AND violation IS NOT NULL AND stop_duration IS NOT NULL
) AS age_violation_data
GROUP BY age_group, violation, stop_duration
ORDER BY age_group, violation, stop_duration;
"""

# Execute and fetch
mycursor.execute(query)
results = mycursor.fetchall()

# Display as table
table = PrettyTable()
table.field_names = ["Age Group", "Violation", "Stop Duration", "Count"]

for row in results:
    table.add_row(row)

print(table)
```

```python
# 5.Violations with High Search and Arrest Rates (Window Function)

query = """
SELECT 
    violation,
    total_stops,
    total_searches,
    total_arrests,
    ROUND((total_searches / total_stops) * 100, 2) AS search_rate_percent,
    ROUND((total_arrests / total_stops) * 100, 2) AS arrest_rate_percent,
    RANK() OVER (ORDER BY (total_searches / total_stops) DESC) AS search_rank,
    RANK() OVER (ORDER BY (total_arrests / total_stops) DESC) AS arrest_rank
FROM (
    SELECT 
        violation,
        COUNT(*) AS total_stops,
        SUM(CASE WHEN search_conducted = TRUE THEN 1 ELSE 0 END) AS total_searches,
        SUM(CASE WHEN is_arrested = 1 THEN 1 ELSE 0 END) AS total_arrests
    FROM traffic_stops
    WHERE violation IS NOT NULL
    GROUP BY violation
) AS violation_stats
ORDER BY search_rate_percent DESC;
"""

# Execute and fetch
mycursor.execute(query)
results = mycursor.fetchall()

# Display with PrettyTable
table = PrettyTable()
table.field_names = [
    "Violation", "Stops", "Searches", "Arrests", 
    "Search %", "Arrest %", "Search Rank", "Arrest Rank"
]

for row in results:
    table.add_row(row)

print(table)
```

```python
# 6 Driver Demographics by Country (Age, Gender, and Race)

query = """
SELECT 
    country_name,
    driver_gender AS gender,
    driver_race AS race,
    CASE 
        WHEN driver_age BETWEEN 18 AND 25 THEN '18-25'
        WHEN driver_age BETWEEN 26 AND 35 THEN '26-35'
        WHEN driver_age BETWEEN 36 AND 45 THEN '36-45'
        WHEN driver_age BETWEEN 46 AND 60 THEN '46-60'
        WHEN driver_age > 60 THEN '60+'
        ELSE 'Unknown'
    END AS age_group,
    COUNT(*) AS total_drivers
FROM traffic_stops
WHERE driver_age IS NOT NULL
GROUP BY country_name, gender, race, age_group
ORDER BY country_name, gender, race, age_group;
"""

# Execute the query
mycursor.execute(query)

# Fetch the results
results = mycursor.fetchall()

# Create a PrettyTable object and set the column names
table = PrettyTable()
table.field_names = ["Country", "Gender", "Race", "Age Group", "Total Drivers"]

# Add each result row to the table
for row in results:
    table.add_row(row)

# Print the table
print(table)

```

```python
# 7.Top 5 Violations with Highest Arrest Rates

query = """
SELECT 
    violation,
    COUNT(*) AS total_stops,
    SUM(CASE WHEN is_arrested = 1 THEN 1 ELSE 0 END) AS total_arrests,
    ROUND((SUM(CASE WHEN is_arrested = 1 THEN 1 ELSE 0 END) / COUNT(*)) * 100, 2) AS arrest_rate
FROM traffic_stops
GROUP BY violation
ORDER BY arrest_rate DESC
LIMIT 5;
"""

# Execute the query
mycursor.execute(query)

# Fetch the results
results = mycursor.fetchall()

# Create and display the table in a readable format using PrettyTable
table = PrettyTable()
table.field_names = ["Violation", "Total Stops", "Total Arrests", "Arrest Rate (%)"]

# Add each result row to the table
for row in results:
    table.add_row(row)

# Print the table
print(table)

```
